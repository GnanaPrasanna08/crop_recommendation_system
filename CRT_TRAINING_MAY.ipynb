```python
class Solution:
    def sortString(self, s: str) -> str:
        ans = ""
        while len(ans) < len(s):
            for ch in sorted(set(s)):
                ans += ch
                s = s.replace(ch, "", 1)
            for ch in sorted(set(s), reverse=True):
                ans += ch
                s = s.replace(ch, "", 1)
        return ans
```

```python
def LinearSearch(arr, key):
  for i in range(len(arr)):
    if arr[i] == key:
      return i
  else:
    return -1
print(LinearSearch([1,2,3,4], 5))
```

```python
def BinarySearch(l, key):
  start = 0
  stop = len(l)-1    #4
  while start <= stop:
    mid = (start + stop)//2
    if l[mid] == key:
      return mid
    elif l[mid] < key:
      start = mid + 1
    else:
      stop = mid - 1
  return -1
print(BinarySearch([1,2,3,4,5,6],6 ))
```

```python
#Bubble Sort => n*(n-1)/2 [no. of comparisons]
a = [3,5,7,2]
n = len(a)
for i in range(n):
  for j in range(0,n-i-1):
    if a[j] > a[j+1]:
      a[j], a[j+1] = a[j+1], a[j]
print(a)
```

```python
#Insertion Sort
l = [3,2,5,1]
n = len(l)
for i in range(1, n):
  key = l[i]
  j = i - 1
  while(j>=0) and (l[j]>key):
    l[j+1] = l[j]
    j = j - 1
  l[j+1] = key
print(l)
```

```python
#Selection Sort
arr = [3,5,7,2]
for i in range(len(arr)):
  min = i
  for j in range(i+1, len(arr)):
    if arr[j] < arr[min]:
      min = j
    arr[i], arr[min] = arr[min], arr[i]
print(arr)
```

```python
#Merge Sort
def MergeSort(arr):
  if len(arr) > 1:
    lefta = arr[:len(arr)//2]
    righta = arr[len(arr)//2:]
    MergeSort(lefta)
    MergeSort(righta)
    i = j = k = 0
    while i < len(lefta) and j < len(righta):
      if lefta[i] < righta[j]:
        arr[k] = lefta[i]
        i += 1
      else:
        arr[k] = righta[j]
        j += 1
      k += 1
    while i < len(lefta):
      arr[k] = lefta[i]
      i += 1
      k += 1
    while j < len(righta):
      arr[k] = righta[j]
      j+=1
      k+=1
    return arr
print(MergeSort([5,5,5,3,8,4]))
```

```python
#Quick Sort
def QuickSort(arr):
  if len(arr) <= 1:
    return arr
  left = []
  right = []
  pvt = arr[0]
  for i in range(1, len(arr)):
    if arr[i] < pvt:
      left.append(arr[i])
    else:
      right.append(arr[i])
  return QuickSort(left) + [pvt] + QuickSort(right)
print(QuickSort([5,5,5,3,8,4]))
```

```python
#Quick Sort
def QuickSort(arr):
  if len(arr) <= 1:
    return arr
  left = []
  right = []
  n = len(arr)//2
  pvt = arr[n]
  for i in range(len(arr)):
    if arr[i] == pvt:
      continue
    if arr[i] < pvt:
      left.append(arr[i])
    else:
      right.append(arr[i])
  return QuickSort(left) + [pvt] + QuickSort(right)
print(QuickSort([5,5,5,3,8,4]))
```
